```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/Salary Data.csv")
```

```python
df.head()
```

```python
def categorize_age(age):
 if 20 <= age < 30:
   return '20-30'
 elif 30 <= age < 40:
   return '30-40'
 elif 40 <= age < 50:
   return '40-50'
 elif 50 <= age < 60:
   return '50-60'
 else:
   return 'Unknown'
```

```python
df['AgeGroup'] = df['Age'].apply(categorize_age)
```

```python
statistics = df.groupby('AgeGroup')['Salary'].describe()
```

```python
statistics
```

```python
df.groupby('AgeGroup')['Salary'].median()
```

```python
df.groupby('AgeGroup')['salary'].apply(lambda x: x.mode().iloc[0])
```

```python
max_age=df.groupby('AgeGroup')['Salary'].max()
max_age
```

```python
min_age=df.groupby('AgeGroup')['Salary'].min()
min_age
```

```python
range=(max_age)-(min_age)
range
```

```python
mid_range=range/2
mid_range
```

```python
df.groupby('AgeGroup')['Salary'].var()
```

```python
df.groupby('AgeGroup')['Salary'].std()
```
